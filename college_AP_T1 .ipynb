```python
# 3. Basic Dataset Analysis
def analyze_dataset(df):
    """
    Performs basic analysis on the dataset.

    Parameters:
    - df (DataFrame): The DataFrame to analyze.
    """
    print("Dataset Overview:")
    print(df.info())
    print("\nBasic Statistics:")
    print(df.describe())
    print("\nMissing Values:")
    print(df.isnull().sum())
    
    # Visualizations
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x='GENDER', hue='ALLOCATED CATEGORY')
    plt.title('Gender vs Allocated Category')
    plt.show()

    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x='REGION', hue='Group')
    plt.title('Region vs Group')
    plt.show()
```
